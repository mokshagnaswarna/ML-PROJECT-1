```python
x=df.drop(columns='math_score',axis=1)
x.head()
```

```python
#create coloumn transformer with 3 types of transformers
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.pipeline import Pipeline
#categorical features
cat_features=[feature for feature in x.columns if x[feature].dtype=='object']
#numerical features
num_features=[feature for feature in x.columns if x[feature].dtype!='object']
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder()
preprocessor=ColumnTransformer(
    [
        ('onehotencoder',oh_transformer,cat_features),
        ('standardscaler',numeric_transformer,num_features)
    ]
)
```

```python
x=preprocessor.fit_transform(x)
```

```python
y=df['math_score']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape
```

```python
def evaluate_model(true,pred):
    mae=mean_absolute_error(true,pred)
    mse=mean_squared_error(true,pred)
    rmse=np.sqrt(mse)
    r2_square=r2_score(true,pred)
    return mae,mse,rmse,r2_square
```

```python

models={
    'Linear Regression':LinearRegression(),
    'Random Forest':RandomForestRegressor(),
    'Gradient Boosting':GradientBoostingRegressor(),
    'XGBRegressor':XGBRegressor(),
    'CatBoosting Regressor':CatBoostRegressor(verbose=False),
    'KNN Regressor':KNeighborsRegressor(),
    'Decision Tree Regressor':DecisionTreeRegressor(),
    'SVR':SVR(),
    'Ridge Regression':Ridge(),
    'Lasso Regression':Lasso()
}
model_list=[]
for i in models:
    model=models[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    mae,mse,rmse,r2_square=evaluate_model(y_test,y_pred)
    print(f'{i} performance:')
    print(f'MAE: {mae}')
    print(f'MSE: {mse}')
    print(f'RMSE: {rmse}')
    print(f'R2 Square: {r2_square}')
    print('-----------------------------')
```
